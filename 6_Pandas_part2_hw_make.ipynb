## Задание 2.3

Добавьте в таблицу новую колонку `Weekday` – день недели на русском языке, т.е. строку со значением `u'понедельник'`, `u'вторник'` и т.д. (*Hint: удобно воспользоваться атрибутом* `.dt`)

```python
df2 = pd.read_excel('data.xlsx', sheetname='curr')
df2['Weekday_num'] =df2.Date.dt.weekday
df2
```

```python
def Weekday(Weekday_num):
    if Weekday_num == 1:
        return 'Вторник'
    elif Weekday_num == 2:
        return 'Среда'
    elif Weekday_num == 3:
        return 'Четверг'
    elif Weekday_num == 4:
        return 'Пятница'
    elif Weekday_num == 5:
        return 'Суббота'
    elif Weekday_num == 6:
        return 'Воскресенье'
    elif Weekday_num == 7:
        return 'Понедельник'
```

```python
df2['Weekday'] =df2.Weekday_num.apply(Weekday)
```
